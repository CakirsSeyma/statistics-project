In this cell, it finds the length of the list and returns it.

```python
def lengthOfList(liste):
    numberOfItem = 0
    for i in liste:
        numberOfItem+=1
    return numberOfItem    
```

In this cell, it finds the sum of the list and returns it.

```python
def sumOfList(liste):
    sum = 0
    for i in liste:
        sum += i
    return sum    
```

I return the average using the sumOfList() function and the length() function

```python
def findMean(liste):
    return sumOfList(liste) / lengthOfList(liste)
```

```python
print(f" Mean of 2018: {findMean(liste_2018)}")
print(f" Mean of 2019: {findMean(liste_2019)}")
```

To find the median, I first sort my list with the sorted() function. If my list is even in length, I return the average of the middle two elements. If it's an odd number, I'm returning the middle element. This is our median value.

```python
def fingMedian(liste):
    sortedList = sorted(liste)
    if lengthOfList(sortedList) % 2 == 0:
       median = (sortedList[lengthOfList(sortedList) // 2 - 1] + sortedList[lengthOfList(sortedList) // 2]) / 2
    else:
       median = sortedList[lengthOfList(sortedList) // 2]
    return median
```

```python
print(f"Median of 2018: {fingMedian(liste_2018)}")
print(f"Median of 2019: {fingMedian(liste_2019)}")
```

I calculate the variance value in this cell
I'm using a loop to calculate the square of the mean and the difference of each element in the array and assign it to the variable 'sqDiff'. Finally, I divide sqDiff value by size.


```python
def calculateVariance(array, size, mean):
    sqDiff = 0
    for i in range(0,size):
        sqDiff += (array[i] - mean) **2 
    variance = sqDiff / size
    return variance
```

```python
print(f"Variance of 2018: {calculateVariance(liste_2018, lengthOfList(liste_2018),findMean(liste_2018))}")
print(f"Variance of 2019: {calculateVariance(liste_2019, lengthOfList(liste_2019),findMean(liste_2019))}")
```

```python
def standartDeviation(array, size, mean):
    return calculateVariance(array, size, mean) ** (1/2)
```

```python
print(f"Standart Deviation of 2018: {standartDeviation(liste_2018, lengthOfList(liste_2018),findMean(liste_2018))}")
print(f"Standart Deviation of 2019: {standartDeviation(liste_2019, lengthOfList(liste_2019),findMean(liste_2019))}")
```

```python
def standartError(standartSapma, size):
    return (standartSapma / (size)**2)
```

```python
print(f"Standart Error of 2018 is : {standartError(standartDeviation(liste_2018, lengthOfList(liste_2018),findMean(liste_2018)),lengthOfList(liste_2018))}" )
print(f"Standart Error of 2019 is : {standartError(standartDeviation(liste_2019, lengthOfList(liste_2019),findMean(liste_2019)),lengthOfList(liste_2019))}" )
```

A z-score is calculated for each value in the data array. The Z-score expresses the position of a value around the mean in standard deviations. Values whose Z-score is greater or less than a certain threshold value (1 by default) are considered outliers. These outliers are added to a list called 'outliers'.

```python
def removeOutliers(data, threshold=1):
    mean = np.mean(data)
    std = np.std(data)
    outliers = []
    
    for i in data:
        z_score = (i - mean) / std
        if np.abs(z_score) > threshold:
            outliers.append(i)
    
    dataWithoutOutliers = [x for x in data if x not in outliers]
    
    return dataWithoutOutliers

dataWithoutOutliers = removeOutliers(liste_2018)
dataWithoutOutliers
```

I take enough data to sample from the list and calculate its mean and standard deviation. I calculate the t value with t.ppf(). I calculate the value for 95% confidence. I do the calculations for the lower and upper limit values. The reason why I use t score is, while doing research, I saw that it would be more appropriate to use t score if the size of our data set is small.

```python
def confidenceInterval(liste, numberOfSample):
    
    sample = np.random.choice(liste , size=numberOfSample, replace=False)

    sampleMean = findMean(sample)
    sampleStandartDeviation = standartDeviation(sample,numberOfSample,sampleMean)
    tValue = t.ppf(0.975, df=numberOfSample-1) #95 CI
    
    lowerBound = sampleMean - tValue * (sampleStandartDeviation / np.sqrt(numberOfSample))
    upperBound = sampleMean + tValue * (sampleStandartDeviation / np.sqrt(numberOfSample))
    
    return (lowerBound, upperBound) 
```

```python
interval_2018 = confidenceInterval(liste_2018,5)
interval_2018
```

```python
interval_2019 = confidenceInterval(liste_2019,4)
interval_2019
```

I calculated the alpha value corresponding to the confidence level. Then I used the norm.ppf() library to specify the sample size and thus I used the z score here as well. The Z-score also expresses the percentile point value of the normal distribution, which is determined according to the confidence level and alpha value. For your sample size, I multiplied the z-score by the standard deviation and divided it by the margin value, rounding to the nearest integer.

```python
def calculateSampleSize(confidenceLevel, margin, stdDeviation):
    alpha = 1 - confidenceLevel
    zScore = norm.ppf(1 - alpha / 2)
    sampleSize = math.ceil((zScore * stdDeviation / margin) ** 2)
    return sampleSize
```

```python
sampleSize_2018 = calculateSampleSize(0.90, 0.1, np.std(liste_2018))
print("Required sample size for 2018:", sampleSize_2018)

sampleSize_2019 = calculateSampleSize(0.90, 0.1, standartDeviation(liste_2019,lengthOfList(liste_2019),findMean(liste_2019)))
print("Required sample size for 2019:", sampleSize_2019)
```
